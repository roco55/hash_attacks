```python
import hashlib 
import numpy as np
import random
import string
import statistics
```

```python
randomletter = "abcdefghijklmnopqrstuvwxyz0123456789!№;%:?*()_-+=&#@"
message = "pochynokyuriisergiyovuch"
```

```python
def hashing(message):
    my_hash = hashlib.blake2b(message.encode()).hexdigest()
    return my_hash
```

```python
def random_message():
    message = "pochynok"
    for i in range(random.randint(0,9)):
        message += random.choice(randomletter)
    message += "yurii"
    for i in range(random.randint(0,9)):
        message += random.choice(randomletter)
    message += "sergiyovuch"
    return message
```

```python
def add_number(message):
    temp = random.randint(0,9)
    new_message = message+str(temp)
    return new_message

def random_but_not_the_samge(message):
    lenght = len(message)
    temp = random.randint(0,lenght-1)
    chooshenone = random.choice(randomletter)
    while(message[temp] == chooshenone):
        chooshenone = random.choice(randomletter)
    return chooshenone, temp


def change_message(message):
    change, temp = random_but_not_the_samge(message)
    message = message[:temp] + change + message[temp+1:]
    return message
```

```python
def attack1(message):
    hash1 = hashing(message)
    new_message = message + "1"
    hash2 = hashing(new_message)
    i = 1
    while(hash1[-4:] != hash2[-4:]):
        i += 1
        new_message = message + f"{i}"
        hash2 = hashing(new_message)
    print("message =", message)
    print("Fixed hash =", hash1, "||| collision =", hash1[-4:])
    print("message with collision =", new_message)
    print(f"hash_{i} with collision =", hash2, "||| collision =", hash2[-4:])
    return i

def attack2(message):
    hash1 = hashing(message)
    new_message = change_message(message)
    hash2 = hashing(new_message)
    i = 0
    while(hash1[-4:] != hash2[-4:] or new_message == message):
        i += 1
        new_message = change_message(new_message)
        hash2 = hashing(new_message)
    print("message =", message)
    print("Fixed hash =", hash1, "||| collision =", hash1[-4:])
    print("message with collision =", new_message)
    print(f"hash_{i} with collision =", hash2, "||| collision =", hash2[-4:])
    return i
```

```python
def happy_birthday(message):
    hash1 = hashing(message)
    new_message = message + "1"
    hash2 = hashing(new_message)
    dict = {}
    dict[hash1[-8:]] = message
    i = 1
    while(hash2[-8:] not in dict.keys()):
        i += 1
        dict[hash2[-8:]] = new_message
        new_message = message + f"{i}"
        hash2 = hashing(new_message)
    print(f"hash_{len(dict)} =", hash2, "||| collision =", hash2[-8:])
    print(f"message_{len(dict)} =", message+f"{len(dict)}")
    print(f"hash_{list(dict).index(hash2[-8:])} =", hashing(dict[hash2[-8:]]), "||| collision =", hashing(dict[hash2[-8:]])[-8:])
    print(f"message_{list(dict).index(hash2[-8:])} =", message+f"{list(dict).index(hash2[-8:])}")
    return i
        
def happy_birthday_2(message):
    hash1 = hashing(message)
    new_message = change_message(message)
    hash2 = hashing(new_message)
    dict = {}
    dict[hash1[-8:]] = message
    i = 1
    while((hash2[-8:] not in dict.keys()) or (new_message in dict.values())):
        i += 1
        dict[hash2[-8:]] = new_message
        new_message = change_message(new_message)
        hash2 = hashing(new_message)
    print(f"hash_{len(dict)} =", hash2, "||| collision =", hash2[-8:])
    print(f"message_{len(dict)} =", new_message)
    print(f"hash_{list(dict).index(hash2[-8:])} =", hashing(dict[hash2[-8:]]), "||| collision =", hashing(dict[hash2[-8:]])[-8:])
    print(f"message_{list(dict).index(hash2[-8:])} =", dict[hash2[-8:]])
    return  i
```

```python
def hundred_tries():
    prob1 = []
    prob2 = []
    col1 = []
    col2 = []
    for i in range(100):
        prob1.append(attack1(random_message()))
        prob2.append(attack2(random_message()))
        col1.append(happy_birthday(random_message()))
        col2.append(happy_birthday_2(random_message()))
    return prob1,prob2,col1,col2

X1,X2,X3,X4 = hundred_tries()
```
